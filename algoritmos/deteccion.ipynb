# Operacioens realizadas para la detección de regiones brillantes.
- 1. Cargar imagen a color
- 2. Convertir a escala de grises
- 3. Suavizado gaussiano para reducir ruido  GaussianBlur
- 4. Umbral adaptativo para detectar zonas brillantes adaptiveThreshold
- 5. Operaciones morfológicas (apertura + cierre).

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.measure import label, regionprops

# Lista de imágenes originales
imagenes = ['../image/1.png', '../image/2.png', '../image/3.png', '../image/4.png']

for i, path in enumerate(imagenes):
    # 1. Cargar imagen a color
    img_color = cv2.imread(path)
    if img_color is None:
        print(f"⚠️ No se pudo cargar {path}")
        continue

    # 2. Convertir a escala de grises
    gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)

    # 3. Suavizado gaussiano para reducir ruido
    blurred = cv2.GaussianBlur(gray, (5,5), 0)

    # 4. Umbral adaptativo para detectar zonas brillantes
    binary = cv2.adaptiveThreshold(blurred, 255,
                                   cv2.ADAPTIVE_THRESH_MEAN_C,
                                   cv2.THRESH_BINARY,
                                   35,  # tamaño del bloque
                                   -10) # constante de ajuste

    # 5. Operaciones morfológicas (apertura + cierre)
    kernel = np.ones((5,5), np.uint8)
    morph = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)
    morph = cv2.morphologyEx(morph, cv2.MORPH_CLOSE, kernel)

    # 6. Etiquetado de regiones
    labeled = label(morph)
    props = regionprops(labeled)

    # Copia en color para dibujar resultados
    result = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

    print(f"\n===== Imagen {i+1} =====")
    for j, region in enumerate(props, 1):
        if region.area >= 100:  # filtrar objetos pequeños (ruido)
            y0, x0 = region.centroid
            minr, minc, maxr, maxc = region.bbox

            # Dibujar bounding box
            cv2.rectangle(result, (minc, minr), (maxc, maxr), (0,0,255), 2)

            # Dibujar centroide
            cv2.circle(result, (int(x0), int(y0)), 5, (0,255,0), -1)

            # Mostrar en consola propiedades
            print(f"Lámpara {j}: Área={region.area}, Perímetro={region.perimeter:.2f}, "
                  f"Centroide=({x0:.1f}, {y0:.1f})")

    # 7. Mostrar resultados
    fig, axs = plt.subplots(1,3, figsize=(14,5))
    axs[0].imshow(gray, cmap='gray')
    axs[0].set_title("Escala de grises")
    axs[0].axis("off")

    axs[1].imshow(morph, cmap='gray')
    axs[1].set_title("Máscara de regiones brillantes")
    axs[1].axis("off")

    axs[2].imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
    axs[2].set_title("Detección de lámparas")
    axs[2].axis("off")

    plt.tight_layout()
    plt.show()

    # 8. Guardar resultado
    cv2.imwrite(f"../image/resultados/lamparas_detectadas_{i+1}.png", result)
```
